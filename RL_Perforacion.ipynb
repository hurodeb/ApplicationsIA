```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('bmh')
```

```python
data = pd.read_csv('datapervolr.csv')
```

```python
data1 = data[['drilling_time','rpm','pull_down','material','depthm']]
```

```python
data1 = data1.dropna()
```

```python
plt.figure(figsize=(20,5),dpi=100)
plt.subplot(1,4,1)
sns.boxplot(data1['rpm'],orient="v")
plt.subplot(1,4,2)
sns.boxplot(data1['depthm'],orient="v")
plt.subplot(1,4,3)
sns.boxplot(data1['pull_down'],orient="v")
plt.subplot(1,4,4)
sns.boxplot(data1['drilling_time'],orient="v")
```

```python
# rpm #

cQ1 = data1['rpm'].quantile(0.25)
cQ3 = data1['rpm'].quantile(0.75)
cIQR = cQ3-cQ1
cBI_Calculado = (cQ1-1.5*cIQR)
cBS_Calculado = (cQ3+1.5*cIQR)

# depth #

dQ1 = data1['depthm'].quantile(0.25)
dQ3 = data1['depthm'].quantile(0.75)
dIQR = dQ3-dQ1
dBI_Calculado = (dQ1-1.5*dIQR)
dBS_Calculado = (dQ3+1.5*dIQR)

# pull_down #

eQ1 = data1['pull_down'].quantile(0.25)
eQ3 = data1['pull_down'].quantile(0.75)
eIQR = eQ3-eQ1
eBI_Calculado = (eQ1-1.5*eIQR)
eBS_Calculado = (eQ3+1.5*eIQR)

# time_per #

fQ1 = data1['drilling_time'].quantile(0.25)
fQ3 = data1['drilling_time'].quantile(0.75)
fIQR = fQ3-fQ1
fBI_Calculado = (fQ1-1.5*fIQR)
fBS_Calculado = (fQ3+1.5*fIQR)
```

```python
data1_so=((data1['rpm']>= cBI_Calculado) & (data1['rpm']<= cBS_Calculado)
          &(data1['depthm']>= dBI_Calculado) & (data1['depthm']<= dBS_Calculado)
          &(data1['pull_down']>= eBI_Calculado) & (data1['pull_down']<= eBS_Calculado)
          &(data1['drilling_time']>= fBI_Calculado) & (data1['drilling_time']<= fBS_Calculado))
          
data2 = data1[data1_so]
data2
```

```python
plt.figure(figsize=(20,5),dpi=100)
plt.subplot(1,4,1)
sns.boxplot(data2['rpm'],orient="v")
plt.subplot(1,4,2)
sns.boxplot(data2['depthm'],orient="v")
plt.subplot(1,4,3)
sns.boxplot(data2['pull_down'],orient="v")
plt.subplot(1,4,4)
sns.boxplot(data2['drilling_time'],orient="v")
```

```python
# rpm #

cQ1 = data2['rpm'].quantile(0.25)
cQ3 = data2['rpm'].quantile(0.75)
cIQR = cQ3-cQ1
cBI_Calculado = (cQ1-1.5*cIQR)
cBS_Calculado = (cQ3+1.5*cIQR)

# depth #

dQ1 = data2['depthm'].quantile(0.25)
dQ3 = data2['depthm'].quantile(0.75)
dIQR = dQ3-dQ1
dBI_Calculado = (dQ1-1.5*dIQR)
dBS_Calculado = (dQ3+1.5*dIQR)

# pull_down #

eQ1 = data2['pull_down'].quantile(0.25)
eQ3 = data2['pull_down'].quantile(0.75)
eIQR = eQ3-eQ1
eBI_Calculado = (eQ1-1.5*eIQR)
eBS_Calculado = (eQ3+1.5*eIQR)

# time_per #

fQ1 = data2['drilling_time'].quantile(0.25)
fQ3 = data2['drilling_time'].quantile(0.75)
fIQR = fQ3-fQ1
fBI_Calculado = (fQ1-1.5*fIQR)
fBS_Calculado = (fQ3+1.5*fIQR)
```

```python
data2_so=((data2['rpm']>= cBI_Calculado) & (data2['rpm']<= cBS_Calculado)
          &(data2['depthm']>= dBI_Calculado) & (data2['depthm']<= dBS_Calculado)
          &(data2['pull_down']>= eBI_Calculado) & (data2['pull_down']<= eBS_Calculado)
          &(data2['drilling_time']>= fBI_Calculado) & (data2['drilling_time']<= fBS_Calculado))
          
data3 = data2[data2_so]
```

```python
plt.figure(figsize=(20,5),dpi=100)
plt.subplot(1,4,1)
sns.boxplot(data3['rpm'],orient="v")
plt.subplot(1,4,2)
sns.boxplot(data3['depthm'],orient="v")
plt.subplot(1,4,3)
sns.boxplot(data3['pull_down'],orient="v")
plt.subplot(1,4,4)
sns.boxplot(data3['drilling_time'],orient="v")
```

```python
materialn = pd.get_dummies(data3['material'],drop_first=True)
```

```python
data3.drop('material',axis=1,inplace=True)
```

```python
data3 = pd.concat([data3,materialn],axis=1)
```

```python
data3.head()
```

```python
X = data3[['rpm','pull_down','depthm','Medio','Muy Duro','Relleno','Suave']]
y = data3['drilling_time']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.figure(figsize=(8,6),dpi=100)
plt.scatter(y_test, predictions)
plt.xlabel ('Tiempo de Perforacin - Test')
plt.ylabel ('Tiempo de Perforacin - Modelo')
plt.xlim(300,2000)
plt.ylim(300,2000)
plt.scatter(y_test,y_test)
```

```python
sns.jointplot(x=y_test, y=predictions,kind='scatter')
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
